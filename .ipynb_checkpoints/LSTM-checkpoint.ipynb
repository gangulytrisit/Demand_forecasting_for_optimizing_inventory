```python
# Function to standardize date formats
def standardize_date(date_str):
    try:
        # Try parsing with the known format "01 December 2021"
        return pd.to_datetime(date_str, format='%d %B %Y')
    except ValueError:
        try:
            # Try parsing with another known format "09-12-2023"
            return pd.to_datetime(date_str, format='%d-%m-%Y')
        except ValueError:
            # If neither format works, return pd.NaT (Not a Time)
            return pd.NaT

# Apply the function to the date columns in both datasets
transactional_data_1['InvoiceDate'] = transactional_data_1['InvoiceDate'].apply(standardize_date)
transactional_data_2['InvoiceDate'] = transactional_data_2['InvoiceDate'].apply(standardize_date)

# Check for rows with NaT (missing/invalid dates)
invalid_dates_1 = transactional_data_1[transactional_data_1['InvoiceDate'].isna()]
invalid_dates_2 = transactional_data_2[transactional_data_2['InvoiceDate'].isna()]

# Display invalid dates if any exist
if not invalid_dates_1.empty:
    print("Invalid dates in Transactional_data_retail_01:")
    print(invalid_dates_1.head())
if not invalid_dates_2.empty:
    print("Invalid dates in Transactional_data_retail_02:")
    print(invalid_dates_2.head())

# Optionally: Drop rows with NaT in the 'InvoiceDate' column
transactional_data_1.dropna(subset=['InvoiceDate'], inplace=True)
transactional_data_2.dropna(subset=['InvoiceDate'], inplace=True)

# Concatenate both datasets into one
transactional_data = pd.concat([transactional_data_1, transactional_data_2], ignore_index=True)
```

```python
import torch
import torch.nn as nn

class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out, (hn, cn) = self.lstm(x)  # LSTM returns the output and the hidden states
        out = self.fc(hn[-1])  # Take the last hidden state
        return out

# Set parameters
input_size = len(top_10_products)  # Number of products
hidden_size = 50  # Number of LSTM units
output_size = len(top_10_products)  # Predicting quantities for the top 10 products

# Create the model
model = LSTMModel(input_size, hidden_size, output_size)
```

```python
# Hyperparameters
num_epochs = 400
learning_rate = 0.001

# Loss and optimizer
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# Convert data to PyTorch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32)
X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val, dtype=torch.float32)

# Training loop
for epoch in range(num_epochs):
    model.train()
    
    # Forward pass
    outputs = model(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)

    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')
```

```python
#validation loss
model.eval()
with torch.no_grad():
    val_outputs = model(X_val_tensor)
    val_loss = criterion(val_outputs, y_val_tensor)

print(f'Validation Loss: {val_loss.item():.4f}')
```
